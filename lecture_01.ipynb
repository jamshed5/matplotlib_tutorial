```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# data frame
data = {'name': ['Tom', 'nick', 'krish', 'jack'],
        'ages': [20, 21, 24, 28],
        'salaries_2001_php': [20000, 22000, 23500, 30000],
        'salaries_2002_php': [20000, 21000, 23000, 28000]}
  
# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
# single plot mean x1 and y1
x1=df.ages
y1=df.salaries_2001_php

# label use for legend
# plot method by default use line
plt.plot(x1,y1, label='salaries_2001_php')

plt.xlabel('ages')
plt.ylabel('salaries_2001_php')
plt.title('salary according to age')

plt.legend()

plt.show()
```

```python
# to get know about style that are available 
plt.style.available
```

```python
# applying style (beauty)
plt.style.use('ggplot')

# multiplot mean x1,y1 and x1,y2
x1=df.ages
y1=df.salaries_2001_php
y2=df.salaries_2002_php

# label use for legend
# for color, can also use code #002200
# applying beauty by using different parameters
plt.plot(x1,y1, color='r', linestyle='--', marker='*', label='salaries_2001_php')
plt.plot(x1,y2, color='#0000FF', marker='o', linewidth='4', label='salaries_2002_php')


plt.xlabel('ages')
plt.ylabel('salaries_2001_php and salaries_2002_php')
plt.title('salary according to age')

plt.legend()

#show grid
plt.grid(True)

# saving plot image
plt.savefig('plot_lecture_01.png')

plt.show()
```
