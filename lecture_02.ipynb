```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
# data frame
data = {'name': ['Tom', 'nick', 'krish', 'jack'],
        'ages': [20, 21, 24, 28],
        'salaries_2001_php': [20000, 22000, 23500, 30000],
        'salaries_2002_php': [20000, 21000, 23000, 28000],
        'salaries_2003_php': [10000, 15000, 20000, 24000],
        'salaries_2004_php': [1000, 5000, 10000, 14000]}
  
# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
# sinmple bar chart
# applying style (beauty)
plt.style.use('ggplot')

x1=df.ages
y1=df.salaries_2001_php

plt.bar(x1,y1, color='b', label='salaries_2001_php')

plt.xlabel('ages')
plt.ylabel('salaries_2001_php')
plt.title('salary according to age')
plt.legend()

#show grid
plt.grid(True)

plt.show()
```

```python
# horizental bar chart
plt.style.use('ggplot')

x1=df.ages
y1=df.salaries_2001_php

# for horizental use barh
plt.barh(x1,y1, color='r', label='salaries_2001_php')

plt.xlabel('salaries_2001_php')
plt.ylabel('ages')
plt.title('salary according to age')
plt.legend()

#show grid
plt.grid(True)

plt.show()
```

```python
# bar chart with multiple data like 2001,2002,2003,2004
# applying style (beauty)
plt.style.use('ggplot')

x_indexes=np.arange(len(df.ages))
print(x_indexes)

width=0.20
print(x_indexes-width)
print(x_indexes+width)
print(x_indexes+2*width)


# multiplot mean x1,y1 and x1,y2
x1=df.ages
y1=df.salaries_2001_php
y2=df.salaries_2002_php
y3=df.salaries_2003_php
y4=df.salaries_2004_php


# beauty is same
# now using x_indexes for x-axies
plt.bar(x_indexes-width,y1, width=width, color='r', label='salaries_2001_php')
plt.bar(x_indexes,y2, width=width, color='b', label='salaries_2002_php')
plt.bar(x_indexes+width,y3, width=width, color='#000000',label='salaries_2003_php')
plt.bar(x_indexes+2*width,y4, width=width, color='#d580ff',label='salaries_2004_php')




# coverting x_indexes to ages on chart
# x1 is ages
plt.xticks(ticks=x_indexes, labels=x1)

plt.xlabel('ages')
plt.ylabel('salaries_2001_php,salaries_2002_php,salaries_2003_php,salaries_2004_php,')
plt.title('salary according to age')

plt.legend()

#show grid
plt.grid(True)

# saving plot image
# plt.savefig('plot_lecture_01.png')

plt.show()
```

```python
# second dataframe
data_2 = {'languages': ['c++', 'c++', 'c++', 'java','java','java','java','java','java','java','java','java','python','python','python','python']}
df2 = pd.DataFrame(data_2)
df2=df2.groupby('languages').value_counts().reset_index()
type(df2)
df2.rename(columns={'languages':'lang',0:'counts'}, inplace=True)
df2
```

```python
# horizental bar chart
plt.style.use('ggplot')

x1=df2.lang
y1=df2.counts

# for horizental use barh
plt.barh(x1,y1, color='r', label='languages')

plt.xlabel('count of languages')
plt.ylabel('languages')
plt.title('languages trend')
plt.legend()

#show grid
plt.grid(True)

plt.show()
```
