```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# data frame
data = {'name': ['Tom', 'nick', 'krish', 'jack','smith'],
        'ages': [20, 21, 24, 28, 30],
        'salaries_2001_php': [20000, 22000, 23500, 30000, 35000],
        'salaries_2002_php': [20000, 21000, 23000, 28000, 35000],
        'salaries_2004_php': [1000, 5000, 10000, 14000, 35000]}
  
# Create DataFrame
df = pd.DataFrame(data)
df
```

```python
df.salaries_2001_php.median()
```

```python
# applying style (beauty)
plt.style.use('ggplot')

# multiplot mean x1,y1 and x1,y2
x1=df.ages
y1=df.salaries_2001_php
y2=df.salaries_2004_php

# label use for legend
# for color, can also use code #002200
# applying beauty by using different parameters
plt.plot(x1,y1, color='red',label='salaries_2001_php')
plt.plot(x1,y2, color='blue',label='salaries_2004_php')


plt.xlabel('ages')
plt.ylabel('salaries_2001_php and salaries_2004_php')
plt.title('salary according to ages')

# only filling for (y1<point) where y1 is salaries_2001_php and median is 23500
median=23500

# for <= less than or equal 'bottom fill'
plt.fill_between(x1,y1, alpha=0.25, where= (y1<=median), label='less than equal <= median salary')
plt.legend()

# for > greate than 'bottom fill'
plt.fill_between(x1,y1, alpha=0.25, where= (y1>median), label='less than > median salary')
plt.legend()



#show grid
plt.grid(True)

# saving plot image
# plt.savefig('plot_lecture_01.png')

plt.show()
```
